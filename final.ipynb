```python
import os
import pandas as pd
import zipfile
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
import torch
from tqdm import tqdm

# ====== YOUR LOCAL PATHS ======
TRAIN_CSV = r"C:\Users\divan\Downloads\dataset\x_train_fixed.csv"
Y_TRAIN_CSV = r"C:\Users\divan\Downloads\dataset\y_train.csv"
TEST_CSV = r"C:\Users\divan\Downloads\test_dataset\x_test.csv"
IMAGE_FOLDER_ZIP = r"C:\Users\divan\Downloads\all_images.zip"
OUTPUT_DIR = r"C:\Users\divan\Downloads\flan_t5_results"

# Create output directory if not exists
os.makedirs(OUTPUT_DIR, exist_ok=True)

# ====== UNZIP IMAGES ======
image_output_path = os.path.join(OUTPUT_DIR, "images")
if not os.path.exists(image_output_path):
    with zipfile.ZipFile(IMAGE_FOLDER_ZIP, 'r') as zip_ref:
        zip_ref.extractall(image_output_path)
    print(f" Extracted images to: {image_output_path}")
else:
    print("ℹ Images already extracted.")

# ====== LOAD DATA ======
train_df = pd.read_csv(TRAIN_CSV)
y_train_df = pd.read_csv(Y_TRAIN_CSV)
test_df = pd.read_csv(TEST_CSV)

# Merge training data with summaries
train_df = pd.merge(train_df, y_train_df, on="sample_id", how="inner")

assert 'noisy_title' in train_df.columns and 'summary' in train_df.columns, \
    "Training CSV must have 'noisy_title' and 'summary' columns after merging."

print(f" Train shape: {train_df.shape}")
print(f" Test shape: {test_df.shape}")

# ====== LOAD MODEL & TOKENIZER ======
model_name = "google/flan-t5-large"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
print(f" Model loaded on: {device}")

# ====== FEW-SHOT EXAMPLES FOR PROMPTING ======
few_shot_examples = train_df.sample(3, random_state=42)

def make_prompt(title, examples):
    prompt = "Generate a concise 2–3 step recipe for the given dish title.\n\n"
    for _, row in examples.iterrows():
        prompt += f"Title: {row['noisy_title']}\nRecipe: {row['summary']}\n\n"
    prompt += f"Title: {title}\nRecipe:"
    return prompt

# ====== INFERENCE ======
results = []
print("\nStarting inference on test set...\n")

for idx, row in enumerate(test_df.itertuples(index=False), 1):
    title = getattr(row, 'noisy_title', getattr(row, 'title', ''))
    image_path = getattr(row, 'image_path', '')

    print(f"Processing {idx}/{len(test_df)}: {title}")

    prompt = make_prompt(title, few_shot_examples)

    inputs = tokenizer(prompt, return_tensors="pt", truncation=True, padding=True).to(device)

    output = model.generate(
        **inputs,
        max_new_tokens=100,
        do_sample=True,
        temperature=0.7,
        top_k=50,
        top_p=0.95
    )

    summary = tokenizer.decode(output[0], skip_special_tokens=True)

    results.append({
        "image_path": image_path,
        "title": title,
        "generated_summary": summary
    })

# ====== SAVE RESULTS ======
results_df = pd.DataFrame(results)
out_csv_path = os.path.join(OUTPUT_DIR, "flan_t5_test_results.csv")
results_df.to_csv(out_csv_path, index=False)
print(f"\nSummarization completed. Results saved to: {out_csv_path}")
```
